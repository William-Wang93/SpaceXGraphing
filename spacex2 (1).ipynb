```python
#Read Data into Pandas
data = pd.read_csv('SpaceXDate.csv')
data_df = pd.DataFrame(data)
```

```python
#Select only the Year from Launch Date column, put it in separate column for analysis
data_df['Launch_Date']=data_df['Launch Date']
data_df['Launch Year'] = data_df.Launch_Date.str[-4:]
#Drop data if no payloard orbit
#data_df.dropna(subset=['Payload Mass (kg)'])
data_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5.5))
sns.set_style("darkgrid")
plt.title('Falcon Payloads Through The Years', size = 16)
ax = sns.scatterplot(x="Launch Year", y="Payload Mass (kg)", data=data, hue = "Launch Site", s= 50)
ax
```

```python
#Create function to assign each site a latitude and longitude
def latitude (row):
   if row['Launch Site'] == 'Marshall Islands' :
      return 7.120520
   if row['Launch Site'] == 'Cape Canaveral AFS LC-40':
      return 28.474009
   if row['Launch Site'] == 'Vandenberg AFB SLC-4E':
      return 34.699460
   if row['Launch Site']  == 'Kennedy Space Center LC-39A':
      return 28.572872

def longitude (row):
   if row['Launch Site'] == 'Marshall Islands' :
      return 171.365753
   if row['Launch Site'] == 'Cape Canaveral AFS LC-40':
      return -80.577171
   if row['Launch Site'] == 'Vandenberg AFB SLC-4E':
      return -120.468150
   if row['Launch Site']  == 'Kennedy Space Center LC-39A':
      return -80.648979
```

```python
#apply latitude and longitude to dataframe
data_df.apply (lambda row: latitude(row), axis=1)
data_df.apply (lambda row: longitude(row), axis=1)
data_df['latitude'] = data_df.apply (lambda row: latitude(row), axis=1)
data_df['longitude'] = data_df.apply (lambda row: longitude(row), axis=1)
fixed_data_df =data_df.rename(columns={"Launch Site": "Launch_Site", "Launch Date": "Launch_Date"})
fixed_data_df
```

```python
count = 0
list = []
for index,row in data_df.iterrows():
    count+= 1
    list.append(count)
l = np.asarray(list)
fixed_data_df['Index']= l
data_df

```

```python
for index,row in fixed_data_df.iterrows():
    fixed_data_df['longitude'] += fixed_data_df['Index']*.00001
fixed_data_df
```

```python
import folium
m = folium.Map(tiles='Stamen Terrain')
for index,row in fixed_data_df.iterrows():
    Launch_Site = row[3]
    Launch_Date = row[1]
    folium.Marker([row[18],row[19]], popup = '<i>Launch_Site:{} Launch_Date:{}<i>'.format(Launch_Site, Launch_Date)).add_to(m)
```
